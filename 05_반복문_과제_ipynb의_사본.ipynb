### 문제2
다음과 같이 두 개의 리스트가 있을 때, zip()을 이용하여 아래와 같이 출력되도록 코드를 작성하세요.

### 문제9
for문과 break를 활용하여,  

1-100까지 숫자중 3과 5의 공배수일 경우 "3과5의공배수",

3의 배수이기만 할 경우 "3의배수",

5의 배수이기만 할 경우 "5의배수",

모두 해당되지 않을 경우 그냥 숫자를 출력하세요.

```python
for i in range(1,101):
    if i % 5 ==0 and i % 3 == 0:
        print("3과 5의 공배수")
    elif i % 3 == 0 :
        print('3의 배수')
    elif i % 5 == 0 :
        print('5의 배수')
    else:
        print(i)
```

### 문제12

### 문제15
다음과 같이 이름, 성별, 나이 데이터가 있는 리스트가 있을 때,

while문을 이용하여 '여자' 데이터를 갖고 있는 사람의 이름을 출력해보세요. 

신원미상인 경우 '(이름)의 성별을 알 수 없습니다'를 출력하고,

break를 사용하여 반복문을 중단하세요.

```python
info = [['김지수', '남자', 24], ['이지현', '여자', 23], ['김보라', '여자', 25],
        ['최수빈', '신원미상', 24], ['김재현', '남자', 22], ['이예진', '여자', 25]]
cnt = 1

while 1:
    if info[cnt-1][1] == '여자':
        print(info[cnt-1][0])
    elif info[cnt-1][1] == '신원미상':
        print(info[cnt-1][0], '의 성별을 알 수 없습니다')
    cnt += 1
    if cnt >= len(info):
        break
```
