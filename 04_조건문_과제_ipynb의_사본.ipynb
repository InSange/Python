### 문제2
이번에는 a가 b보다 크거나 같을 때 'a가 b보다 크거나 같다'를 출력하고, 그렇지 않다면 'b가 a보다 크다'를 출력하는 코드를 작성하세요.

```python
import random

a, b = random.randrange(1,10), random.randrange(1,10)

print(a, b)

if a > b or a == b:
    print('a가 b보다 크거나 같다.')
else:
    print('b가 a보다 크다')
```

### 문제3

```python
num1 = 512
num2 = 512

if num1 > num2:
    print('num1이 더 큽니다')
elif num1 == num2:
    print('두 값이 일치합니다')
else :
    print('num2이 더 큽니다')
```
